```python
def ValidateInput(func):

    def Exec(a,b):

        if a<b:

            op=func(b,a)
        
        else:

            op=func(a,b)
        
        print(op)
    

    return Exec
```

```python

@ValidateInput
def Subtract(a,b):

    return a-b
```

```python
Subtract(63,85)
```

```python
def dec(func):
    
    def time():
        t1 = pd.Timestamp.now()
        func()
        t2 = pd.Timestamp.now()
        time_taken = t2-t1
        print("Time taken in seconds ",time_taken.seconds)

    return time    

@dec
def new():
    for i in range(1,100000000):
        pass
    
result = new()
```

## TIC-TAC-TOE GAME

```python
# CREATING THE GAME

# 1. Welcome the user "Welcome to the game Tic-Tac-Toe"

# 2. SHow EMpty Tic Tac Toe Board

# 3. Ask the 1st user to select the piece "X" or "O"

# 4. Assign the remaining piece to the second user

# 5. From User1 Get the position of his piece

# 6. Place the user's piece in the position

# 7. From User2 Get the position

# 8. Place the user-2's piece in the position

# 9. Show them the board updated with pieces from the user1 and user2

# 10. Check whether game is over or not

# 11. Get input for replay
```

```python
def PrintBoard(board):
    
    # Line 1
    
    print("  ---   ---   ---  ")
    
    print("|  "+board[0]+"  |  "+board[1]+"  |  "+board[2]+"  |")
    
    print("  ---   ---   ---  ")
    
    print("|  "+board[3]+"  |  "+board[4]+"  |  "+board[5]+"  |")
    
    print("  ---   ---   ---  ")
    
    print("|  "+board[6]+"  |  "+board[7]+"  |  "+board[8]+"  |")
    
    print("  ---   ---   ---  ")
    
    
    
def UserSelectPieces(pieces):
    
    "returns user1 piece and user2 piece"
    
    
    p1=input("Please enter 'X' for piece X and 'O' for piece O")
    
    if p1 not in pieces:
        
        p1,p2=UserSelectPieces(pieces)
    
    else:
        
        p2=None
        
        for piece in pieces:
            
            if piece!=p1:
                
                p2=piece
    
    return p1,p2
    


def GetPosition(piece,board):
    
    pos=int(input("Enter the position for your piece from 1 to 9 :"+piece))
    
    if pos not in list(range(1,10)):
        
        print("This is not a valid position")
        
        pos=GetPosition(piece,board)
    
    else:
        
        if board[pos-1]!=" ":
            
            print("Position Already Occupied")
            
            print("Please Enter a valid position")
            
            pos=GetPosition(piece,board)
        
    return pos


def GameOverOrNot(board):
    
    "validated whether the game is over or not"
    
    if " " not in board:
        
        gameOver=True
        
    else:
    
        gameOver=False
    
    winnerPiece=None
    
    winningPos=[(0,1,2),(0,3,6),(0,4,8),(1,4,7),(2,5,8),(3,4,5),(6,7,8)]
    
    for pos in winningPos:
        
        if board[pos[0]]==board[pos[1]] and board[pos[1]]==board[pos[2]] and board[pos[2]]!=" " :
            
            print(pos)
            
            winnerPiece=board[pos[0]]
            
            gameOver=True
        
        
    
    return gameOver,winnerPiece
        
            
            
def TicTacToe():
    
    print("Welcome to the game Tic-Tac-Toe")
    
    board=[" "," "," "," "," "," "," "," "," "]
    
    pieces=["X","O"]
    
    PrintBoard(board)
    
    p1Piece,p2Piece=UserSelectPieces(pieces)    
    
    gameOver=False
    
    winner=None
    
    while gameOver==False:        
        
        p1Pos=GetPosition(p1Piece,board)
        
        board[p1Pos-1]=p1Piece
        
        PrintBoard(board)
        
        gameOver,winner=GameOverOrNot(board)
        
        if gameOver==True:
            
            break
            
        p2Pos=GetPosition(p2Piece,board)
        
        board[p2Pos-1]=p2Piece
        
        PrintBoard(board)
        
        gameOver,winner=GameOverOrNot(board)
        
        if gameOver==True:
            
            break
    
    
    if winner is None:
        
        print("The Game is draw. Nobody Won")
    
    else:
        
        print("Congratulations!!! "+winner+" Won")
    
    
    restart=input("Do you want to restart Yes/No Enter Y or N")
    
    if restart=="Y":
        
        TicTacToe()
    
    else:
        
        print("Bye !!!")
        
        
        
        
    
    
    
    
```
